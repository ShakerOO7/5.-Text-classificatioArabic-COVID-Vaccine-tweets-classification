```python
import snscrape.modules
```

```python
keywords = ['لقاح كورونا', 'لقاح فيروس كورونا', 'كوفيد-19']
```

```python
MAX = 200
tweets_list = []
for keyword in keywords:
    print(keyword)
    if len(tweets_list) > MAX:
        break
    tweets = snscrape.modules.twitter.TwitterSearchScraper(keyword+ ' since:2020-10-15 until:2020-10-17 lang:ar')
    for tweet in tweets.get_items():
        if len(tweets_list) > MAX:
            break
        tweets_list.append(tweet.content.replace('\n', ' ') + '\n')
    print(len(tweets_list))
```

```python
file = open('from 2020-10-15 to 2020-10-17 محمد شاكر اللحام.txt', 'w', encoding="utf-8")
for tweet in tweets_list:
    print(tweet)
```

```python
file.writelines(tweets_list)
```

```python
file.close()
```
